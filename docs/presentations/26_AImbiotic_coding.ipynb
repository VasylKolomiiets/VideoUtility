### Vibe coding. Історія терміну.

Вайб кодінг (від англ. Vibe coding), також відомий як "кодінг на вайбі" або іноді "СДВГ-кодінг" (через нелінійний та інтуїтивний характер процесу), — це новий підхід до розробки програмного забезпечення, який активно обговорюється у 2025 році.

**Визначення та значення:**

В основному, вайб кодінг розуміють як **створення програмного забезпечення за допомогою штучного інтелекту (ШІ) та великих мовних моделей (LLM)**. Замість того, щоб писати код вручну, програміст описує свою ідею або завдання природною мовою, а ШІ генерує відповідний код. Цей підхід дозволяє:

* **Прискорити процес розробки:** ШІ може швидко створювати значні обсяги коду.
* **Спростити розробку для не-програмістів:** Люди без глибоких знань у програмуванні можуть створювати базове програмне забезпечення, описуючи свої потреби.
* **Змінити роль програміста:** Програміст стає більше "оркестратором" процесу, який направляє ШІ та інтегрує згенерований код.

Однак, існує й інше значення цього терміну. У більш широкому сенсі, "вайб кодінг" може описувати **стан програміста, коли код пишеться не заради дедлайнів, а заради задоволення, натхнення та "кайфу" від процесу**. У цьому контексті, це розмова про настрій, творчість та інтуїтивне написання коду.

**Походження терміну:**

Термін "вайб кодінг" став особливо популярним у 2025 році, з розвитком та впровадженням потужних мовних моделей, здатних генерувати код. Вважається, що вперше цей підхід запропонував вчений у галузі штучного інтелекту [Андрій Карпатий](https://karpathy.ai/).

Незважаючи на потенційні переваги, існують і критичні погляди на вайб кодінг. Деякі вважають, що надмірне покладання на ШІ може призвести до втрати глибокого розуміння коду розробниками та створення менш якісного програмного забезпечення. Також є думка, що для ефективного використання вайб кодінгу все ще потрібні навички програміста для розуміння та перевірки згенерованого коду.

Таким чином, "вайб кодінг" - це багатогранний термін, що відображає нову тенденцію в розробці програмного забезпечення, пов'язану з використанням штучного інтелекту, а також може описувати творчий та натхненний стан програміста під час написання коду.

#### Походження самого слова `вайб`  (vibe)

Для того, щоб детальніше розібратися зі словом "vibe" у контексті "вайб кодінг", давайте розглянемо його походження та значення окремо від програмування:

**Походження та значення слова "vibe"**

Слово "vibe" є скороченням від слова "vibration" (вібрація) і вперше з'явилося в англійській мові на початку 20 століття. Його перше вживання фіксується близько 1900 року.

Первинно слово "vibration" використовувалося для опису фізичних коливань. Згодом, у розмовній мові, скорочена форма "vibe" почала використовуватися в переносному значенні для опису:

* **Емоційної атмосфери або настрою місця, події чи ситуації.** Наприклад, "У цього міста особлива творча вайб" або "Вечірка мала дуже позитивну вайб".
* **Інтуїтивного відчуття або враження від людини чи ситуації.** Наприклад, "У мене від цієї людини погана вайб" або "Я відчуваю хорошу вайб від цієї ідеї".

Таким чином, слово "vibe" **не є акронімом**. Воно утворилося шляхом скорочення вже існуючого слова "vibration".

**"Vibe" у контексті "вайб кодінг"**

У контексті "вайб кодінг", слово "vibe" використовується, ймовірно, для передачі ідеї інтуїтивного, неструктурованого та "відчутного" процесу кодування за допомогою ШІ. Програміст ніби "ловитиме вайб" потрібного коду, описуючи його своїми словами, а ШІ на основі цього "відчуття" генеруватиме результат.

Інше значення, як вже згадувалося, підкреслює творчий та натхненний стан програміста, коли він кодує з ентузіазмом, ніби "на хвилі" або "у вайбі" процесу.

Отже, слово "vibe" у "вайб кодінг" походить від розмовного англійського слова, що означає емоційну атмосферу або інтуїтивне відчуття, і не є акронімом.

### Введемо новий термін - ШІмбіотичне кодування.     
     
>
> *Я придумав для процесу парного програмування термін ШІмбіотичне кодування.*
> *Чи є воно тобі зрозумілим без пояснень? Англійською було б `AImbiotic coding`.*
>  
     
Так, термін "ШІмбіотичне кодування" (англійською "AImbiotic coding") є досить зрозумілим без додаткових пояснень, особливо для людей, знайомих з основами біології та технологій.

Ось чому:

* **ШІ (AI):** Ця абревіатура є загальноприйнятою та відразу вказує на використання штучного інтелекту.
* **Симбіотичний (Symbiotic):** У біології термін "симбіоз" описує взаємовигідні відносини між двома різними організмами. Переносячи це поняття в контекст програмування, "ШІмбіотичне кодування" одразу викликає асоціацію з партнерською роботою, де програміст та ШІ співпрацюють, отримуючи взаємну вигоду. Програміст надає завдання, напрямок та експертизу, а ШІ допомагає генерувати код, пропонувати рішення або виконувати рутинні завдання, роблячи процес більш ефективним та продуктивним.

Таким чином, поєднання "ШІ" та "симбіотичний" чітко передає ідею **взаємовигідної співпраці між людиною та штучним інтелектом у процесі програмування**. Термін є влучним і легко сприймається.

>
> _Мені він здається більш влучним для парного програмування аніж вайб кодінг. Так чи ні?_
>

Так, з моєї точки зору, термін "ШІмбіотичне кодування" є більш влучним для опису процесу парного програмування з використанням штучного інтелекту, ніж "вайб кодінг". Ось чому:

* **Підкреслення співпраці:** "ШІмбіотичне кодування" чітко вказує на партнерські, взаємовигідні відносини між програмістом та ШІ, що є ключовою характеристикою парного програмування. Термін "симбіотичний" недвозначно говорить про спільну роботу, де обидві сторони отримують переваги.
* **Точність опису процесу:** Парне програмування, особливо з використанням ШІ, передбачає активну взаємодію та обмін знаннями та можливостями між людиною та машиною. "ШІмбіотичне кодування" добре відображає цю динаміку.
* **Уникнення розмитості:** Термін "вайб кодінг", хоча і може передавати відчуття легкості та інтуїтивності, є більш розмитим і не завжди явно вказує на спільну роботу. Він може стосуватися як індивідуального кодування з ШІ-асистентом, так і власне парного програмування.

Отже, "ШІмбіотичне кодування" є більш конкретним, точним та таким, що краще відображає суть парного програмування з залученням штучного інтелекту, на відміну від більш загального та метафоричного "вайб кодінгу". Ваша назва є досить вдалою та інформативною.

### GitHub Copilot vs Gemini Code Assist з точки зору Gemini

Звичайно, ось порівняння GitHub Copilot та Gemini Code Assist, де спільні можливості розглядаються в одному пункті, а потім наведено перелік унікальних можливостей для кожного:

**Спільні можливості:**

* **Автозавершення коду:**
    * **GitHub Copilot:** Пропонує завершення рядків коду в режимі реального часу під час введення.
    * **Gemini Code Assist:** Автоматично пропонує цілі блоки коду на основі контексту.
* **Генерація коду на основі коментарів:**
    * **GitHub Copilot:** Генерує код, виходячи з опису завдання природною мовою в коментарях.
    * **Gemini Code Assist:** Дозволяє використовувати природну мову для запиту змін у коді.
* **Підтримка багатьох мов програмування:**
    * **GitHub Copilot:** Підтримує широкий спектр популярних мов, таких як Python, JavaScript, TypeScript, Go, Ruby, C++, C#, Java тощо.
    * **Gemini Code Assist:** Також працює з багатьма мовами програмування.
* **Інтеграція з VS Code:**
    * Обидва сервіси безпосередньо інтегруються в середовище розробки Visual Studio Code.

**Унікальні можливості GitHub Copilot:**

* **Глибока інтеграція з GitHub:** Тісно пов'язаний з екосистемою GitHub, що спрощує роботу з репозиторіями та pull requests.
* **Велика спільнота та розвинена документація:** Існує велика кількість користувачів та навчальних матеріалів.
* **Ефективність у генерації шаблонного коду:** Добре справляється з автоматичним створенням стандартних блоків коду.

**Унікальні можливості Gemini Code Assist:**

* **Виявлення помилок у реальному часі:** Ідентифікує потенційні помилки в коді та пропонує варіанти їх виправлення.
* **Локальна обізнаність про кодову базу (у платних версіях):** Має розуміння структури локального проєкту, що може покращити якість пропозицій.
* **Трансформація коду:** Дозволяє використовувати команди для рефакторингу та модифікації існуючого коду з можливістю перегляду змін перед застосуванням.
* **Чат в IDE:** Надає вбудований чат для взаємодії з асистентом, де можна ставити запитання та отримувати пояснення.
* **Розширена інтеграція з Google Cloud (у платних версіях):** Тісна інтеграція з сервісами Google Cloud, такими як Firebase, Colab, Vertex AI та іншими.
